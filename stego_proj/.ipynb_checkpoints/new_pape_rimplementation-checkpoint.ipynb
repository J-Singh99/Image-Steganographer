```python
import numpy as np
from PIL import Image
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def Binary(n): 
    s = bin(n) 
      
    # removing "0b" prefix 
    s1 = s[2:] 
    return s1
def binaryToDecimal(n):
    if n=='':
        n='0'
    return str(int(n,2))
```

```python
def show_pic(img):
    fig=plt.figure(figsize=(10,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
def load_pic(path):
    img=cv2.imread(path)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
```

```python
def resize_pic(img):
    while((img.shape[0]*img.shape[1])%4!=0):
        img=cv2.resize(img,(img.shape[1]-1,img.shape[0]))  #resize function uses (x,y) and shape displays as (y,x) corresponding to height, width and numpy sees as (y,x) as rows and columns
    return img
        
```

```python
def crop_image(original_img,x_off,x_end,y_off,y_end):
    cut_img=original_img[y_off:y_end,x_off:x_end,]
    return cut_img
```
